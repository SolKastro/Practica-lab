```python
import matplotlib.pyplot as plt

x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]
plt.plot(x, y1, label="línea I")
plt.plot(x, y2, label="línea H")
plt.plot()

plt.xlabel("eje x")
plt.ylabel("eje y")
plt.title("Ejemplo de gráfico lineal")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 2, 4, 7, 8, 3]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

plt.show()

plt.bar(x1, y1, label="Cyan", color='c')
plt.bar(x2, y2, label="Verde", color='g')
plt.plot()

plt.xlabel("número de barra")
plt.ylabel("altura de la barra")
plt.title("Ejemplo de gráfico de barras")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.random.randn(1000)

m = [m for m in range(len(n))]
plt.bar(m, n)
plt.title("Datos de calificaciones")
plt.show()

plt.hist(n, bins=20)
plt.title("Histograma")
plt.show()
```

```python
import matplotlib.pyplot as plt

x1 = [2, 3, 4]
y1 = [5, 5, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [2, 3, 2, 3, 4]
y3 = [6, 8, 7, 8, 7]

plt.scatter(x1, y1)
plt.scatter(x2, y2, marker="p", color='r')
plt.scatter(x2, y3, marker='*', color='c')
plt.title('Ejemplo de gráfica de dispersión')
plt.show()
```
